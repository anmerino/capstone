their marketing tactics. Moreno and her team are interested in analyzing
the Cyclistic historical bike trip data to identify trends

### Location of Data

The data used in this project can be found here:
<https://divvy-tripdata.s3.amazonaws.com/index.html> This data has been
collected by Motivate company employed by the city of Chicago to collect
data about the usage of their bikes.

### Data Organisation

The data is stored in csv files that are updated on a monthly basis. For
this research, the most current twelve months of data (January 2021 –
December 2021) were used. The files have 13 columns that contain
information such as ride id, ridership type, trip time, start and end
locations, and geographic coordinates, among other things.

### Credibility of the Data

Motivate, Inc., the corporation that manages the City of Chicago’s
Cyclistic Bike Share programme, collects the data directly. The data is
extensive and includes information from all of the rides recorded on the
system, rather than just a sample. The information is current. It is
published on a monthly basis and was current as of December 2021. The
data is made accessible to the public by the City of Chicago.

### Licensing, Privacy, Security and Accesibility

All identifying information has been removed from this data, making it
anonymous. This protects privacy, but also restricts the scope of the
investigation. There is insufficient data to tell whether casual riders
are repeat riders or if casual riders are Chicago residents. This
licence <https://ride.divvybikes.com/data-license-agreement> governs the
distribution of this data.

### Data’s ability to provide answers to this Business Question

This data is necessary in order to identify differences in how members
or casual riders use the bike share programme identify trends and
difference between them.

### Problems with the dataset

There are some discrepancies in the structure of the data frame. The

solved by data cleansing, but there are a few that require more

More information would be need in order to create a proper

### Data frame with station names, coordinates and type of members.

``` r
trip_stations <- select(cy_trips, start_station_name, end_station_name, start_lat, start_lng, end_lat, end_lng, member_casual)

head(trip_stations)
```

Here I decided to create another row call “Route” to contain the route
follow by the cyclist joining the name of the starting station with the
ending station.

``` r
trip_stations$route <- paste(trip_stations$start_station_name, trip_stations$end_station_name, sep=" TO ")

head(trip_stations) 
```

Created another row to contain the number of times the route was follow
and omit empty ones.

``` r
routes <- trip_stations %>% 
  group_by(route, member_casual, start_lat, start_lng, end_lat, end_lng) %>% 
  count(route, name = "count", sort = TRUE)%>% 
  summarize(route,count) %>%
```



The duration of trips for casual users is almost double that of member
